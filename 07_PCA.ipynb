```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.decomposition import PCA
```

```python
data = load_breast_cancer(as_frame=True)

df = data.frame

df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
X = df.drop('target', axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
X_train.shape
y_test.shape
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

```python
columns= data.feature_names

X_train = pd.DataFrame(X_train, columns=columns)
X_test = pd.DataFrame(X_test, columns=columns)
```

```python
X_train.head()
```

```python
X_test.tail()
```

```python
logistic = LogisticRegression()
gbc = GradientBoostingClassifier()
```

```python
logistic.fit(X_train, y_train)
gbc.fit(X_train, y_train)
```

```python
print(LogisticRegression)

y_pred_logistic = logistic.predict(X_test)

print(f' Accuracy : {accuracy_score(y_test, y_pred_logistic)}\n')
print(f'Confusion Matrix : {confusion_matrix(y_test, y_pred_logistic )}\n')
print(f'Classification Report: {classification_report(y_test, y_pred_logistic)}')

print("--------------------------------------------------------")
print(f'GradientBoostingClassifier:')
y_pred_gbc = gbc.predict(X_test)
print(f'Accuracy Score : {accuracy_score(y_test, y_pred_gbc)}')
print(f'Confusion Matrix : {confusion_matrix(y_test, y_pred_gbc)}')
print(f'Classification Report: {classification_report(y_test, y_pred_gbc)}')
```
